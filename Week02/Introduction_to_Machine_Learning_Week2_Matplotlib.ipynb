```python
import matplotlib.pyplot as plt
import numpy as np

#data

x = np.arange(0,10,0.1) #[0, 0.1, 0.2 .... 9.9]
y = np.sin(x)

plt.plot(x,y)

plt.title("Sinus Grafigi")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Data
x = np.random.rand(50)
y = np.random.rand(50)

colors = np.random.rand(50)
sizes = np.random.randint(50,75,50)

plt.scatter(x, y, c=colors, s=sizes)

plt.title("Nokta Grafigi")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
```

```python
#Data

data = np.random.randn(444)

plt.hist(data, bins=30)

plt.title("Histogram")
plt.xlabel("Degerler")
plt.ylabel("Frekans")
```

```python
#Data

x = ['A','B','C','D']
y = np.random.randint(1,8,4)

plt.bar(x,y)

plt.title("Sutun Grafigi")
plt.xlabel("Kategoriler")
plt.ylabel("Degerler")
```
